## Web based loader

```python
from langchain_community.document_loaders import WebBaseLoader
```

```python
web_loader = WebBaseLoader("https://www.espn.com/")
```

```python
documents = web_loader.load()
print(documents)
```

```python
import bs4

loader = WebBaseLoader(web_paths=["https://www.espn.com/"], bs_kwargs=dict(parse_only=bs4.SoupStrainer(class_="content")))
```

```python
loader.load()
```
