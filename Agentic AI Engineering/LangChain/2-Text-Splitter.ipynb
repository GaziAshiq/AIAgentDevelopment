```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Read the text file
with open("../../Test_Documents/speech.txt", "r") as file:
    text = file.read()

print("Original text length:", len(text))
print("First 200 characters:")
print(text[:200] + "...")
```

```python
# Initialize the text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separators=["\n\n", "\n", " ", ""]
)
```

```python
# Split the text into documents
documents = text_splitter.create_documents([text])

print(f"Number of document chunks: {len(documents)}")
print("\nFirst chunk:")
print(documents[0].page_content)
print(f"\nFirst chunk length: {len(documents[0].page_content)} characters")

if len(documents) > 1:
    print("\nSecond chunk:")
    print(documents[1].page_content)
    print(f"\nSecond chunk length: {len(documents[1].page_content)} characters")
```
